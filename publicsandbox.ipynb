```python
import requests
from dotenv import load_dotenv
import os
load_dotenv()
```

```python

url = "https://api.public.com/userapiauthservice/personal/access-tokens"

payload = {
    "validityInMinutes": 60,
    "secret": os.getenv('PUBLIC_API_KEY')
}

headers = {
    "Content-Type": "application/json"
}

response = requests.post(url, json=payload, headers=headers)
token = response.json().get('accessToken')
```

```python
import requests

url = "https://api.public.com/userapigateway/trading/instruments"
headers = {
    "Authorization": f"Bearer {token}",
    "Content-Type": "application/json"
}
params = {
    "optionTradingFilter": ["BUY_AND_SELL"] ,
    "optionSpreadTradingFilter": ["BUY_AND_SELL"] 
}

response = requests.get(url, headers=headers, params=params)
instruments = response.json()['instruments']
```

```python
import requests

url = "https://api.public.com/userapigateway/trading/account"
headers = {
    "Authorization": f"Bearer {token}",
    "Content-Type": "application/json"
}

response = requests.get(url, headers=headers)
accounts = response.json()['accounts']
brokerageaccnts = [accnt['accountId'] for accnt in accounts if accnt['accountType'] == 'BROKERAGE']
```

```python
import requests

account = brokerageaccnts[0]
url = f"https://api.public.com/userapigateway/marketdata/{account}/option-chain"
headers = {
    "Authorization": f"Bearer {token}",
    "Content-Type": "application/json"
}

request_body = {
  "instrument": {
    "symbol": "AAPL",
    "type": "EQUITY"
  },
  "expirationDate": "2025-09-19"
}


response_options = requests.post(url, headers=headers, json=request_body)
options = response_options.json()
```

```python
options['calls']
```

```python
osiOptionSymbol = 'AAPL250919C00370000'
url = "https://api.public.com/userapigateway/option-details/{account}/{osiOptionSymbol}/greeks"
headers = {
    "Authorization": f"Bearer {token}",
    "Content-Type": "application/json"
}

for option in options['calls']:
    url.format(account=account, osiOptionSymbol=option['instrument']['symbol'] )
    response = requests.get(url, headers=headers)
    data = response.json()
    options['calls']
print(data)
```
