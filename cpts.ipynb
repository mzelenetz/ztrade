```python
import polars as pl
```

```python
df= (pl.read_csv(source="/Users/mz/Documents/projects/ztrade/*92250_cpt.csv", glob=True)
     .with_columns(pl.col("standard_charge_dollar").cast(pl.Float64).alias("standard_charge_dollar"))
     .unique()
     .filter(pl.col("standard_charge_dollar").is_not_null(), pl.col("standard_charge_dollar") < 1000.0)
)
```

```python
import pandas as pd

df = df.to_pandas()
```

```python
import matplotlib.pyplot as plt

# Drop NaNs to avoid matplotlib warnings
vals = df["standard_charge_dollar"].dropna()

plt.figure(figsize=(10, 6))
plt.hist(vals, bins=25)
plt.xlabel("Standard Charge ($)")
plt.ylabel("Frequency")
plt.title("Distribution of Standard Charge For CPT Code 92250 NYC Hospitals")
plt.grid(True)
plt.show()
```
